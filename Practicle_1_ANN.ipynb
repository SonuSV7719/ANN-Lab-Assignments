#Sigmoid Function

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def sigmoid_function(x):
  return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-10, 10)
```

```python
x
```

```python
sns.scatterplot(x)
plt.title("X data")
```

```python
y = sigmoid_function(x)
```

```python
sns.scatterplot(y)
plt.title("Sigmoid Function")
```

```python
plt.plot(y)
plt.title("Sigmoid Function")
```

#Binary function

```python
def binary_function(x):
  return 0 if x<0 else 1
```

```python
binary_output = []
for i in [-10, -8, -5, -2, 0, 1, 2, 3, 8, 10]:
  binary_output.append(binary_function(i))
```

```python
binary_output
```

```python
sns.scatterplot(binary_output)
plt.title("Binary Function scatterplot")
```

```python
plt.plot(binary_output)
plt.title("Binary function line plot")
```

#Tanh

```python
def tanh_function(x):
  return (np.exp(x) - np.exp(-x) / np.exp(x) + np.exp(-x))
```

```python
def tanh_function2(x):
  return 2*sigmoid_function(2*x)-1
```

```python
x = np.linspace(-50, 50, 5000)
```

```python
y = [tanh_function(i) for i in x]
```

#Relu

```python
def relu_function(x):
  return np.array([0, x]).max()
```

```python
x = np.linspace(-10, 10)
```

```python
y = [relu_function(i) for i in x]
```

```python
sns.scatterplot(y)
plt.title("Relu Function")
```

```python
plt.plot(y)
plt.title("Relu Function")
```

#leaky Relu

```python
def leaky_relu_function(x):
  return  0.01*x if x < 0 else x
```

```python
x = np.linspace(-10, 10)
```

```python
y = [leaky_relu_function(i) for i in x]
```

```python
sns.scatterplot(y)
plt.title('Leaky Relu Function')
```

```python
plt.plot(y)
plt.title("Leaky Relu Function")
```
